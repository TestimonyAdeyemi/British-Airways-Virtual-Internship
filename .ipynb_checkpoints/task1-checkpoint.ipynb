```python
import pandas as pd
import numpy as np
import nltk
nltk.download('vader_lexicon')
```

## Data preparation

```python
data = pd.read_csv('dataset.csv')
```

```python
data['review_date'] = data['review_date'].str.replace(r'\b(\d+)(st|nd|rd|th)\b', r'\1', regex=True)
```

```python

# Convert the "review_date" column to datetime
data['review_date'] = pd.to_datetime(data['review_date'], format="%d %B %Y")
```

```python
# Define the start and end date ranges
start_date = pd.to_datetime("2023-01-01")
end_date = pd.to_datetime("2023-10-31")

data1 = data[(data['review_date'] >= start_date) & (data['review_date'] <= end_date)]
```

```python
data1.head()
test = data1[(data1['review_date'] == end_date)]
```

```python
data2 = data1.sort_values(by='review_date')
```

```python
data2['review_month'] = data2['review_date'].dt.month
```

```python
data2.tail()
```

```python
months_name = []
# Iterate through each row in the DataFrame
for index, row in data2.iterrows():
    if row['review_month'] == 1:
        months_name.append("January")
    elif row['review_month'] == 2:
        months_name.append("February")
    elif row['review_month'] == 3:
        months_name.append("March")
    elif row['review_month'] == 4:
        months_name.append("April")
    elif row['review_month'] == 5:
        months_name.append("May")
    elif row['review_month'] == 6:
        months_name.append("June")
    elif row['review_month'] == 7:
        months_name.append("July")
    elif row['review_month'] == 8:
        months_name.append("August")
    elif row['review_month'] == 9:
        months_name.append("September")
    elif row['review_month'] == 10:
        months_name.append("October")

# Create a new column 'months_name' in the DataFrame
data2['month'] = months_name
```

## Average overall rating per month

```python
monthly_overall_rating = data2.groupby('month')['overall_rating'].mean()
monthly_overall_rating.sort_values(ascending=False)
```

```python
custom_colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon', 'lightpink', 'lightseagreen', 'lightblue', 'lightcoral', 'lightsteelblue', 'lightgoldenrodyellow']

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
monthly_overall_rating.plot(kind='bar', color=custom_colors)
plt.title('Monthly Overall Rating')
plt.xlabel('Month')
plt.ylabel('Mean Overall Rating')
plt.xticks(rotation=45)  # Rotate the x-axis labels for readability
plt.show()
```

## Overall Reccomendation rate 

```python
total = data2['recommended'].count()
yes = data2[data2['recommended'] == "yes"]['recommended'].count()
percentage = (yes / total) * 100
percentage
```

```python
labels = ['Yes Recommendations', 'No Recommendations']
sizes = [percentage, 100 - percentage]
colors = ['lightgreen', 'lightcoral']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Percentage of "Yes" Recommendations')
plt.show()
```

## Reccomendation rate per month

```python
result = data2.groupby('month')['recommended'].apply(lambda x: (x == 'yes').sum() / len(x) * 100).reset_index()
result.rename(columns={'recommended': 'percentage_yes'}, inplace=True)
```

```python
result
```

```python
plt.figure(figsize=(10, 6))
plt.bar(result['month'], result['percentage_yes'], color=custom_colors)
plt.title('Percentage of "Yes" Recommendations by Month')
plt.xlabel('Month')
plt.ylabel('Percentage')
plt.xticks(result['month'])
plt.show()
```

## Sentiment Analysis on comments

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
data2['sentiment_score'] = data2['content'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
```

```python
data2.head(5)
```

```python
data2['sentiment_score'].mean()
```

## Wordcloud on comments with positive sentiments

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# targeting comments with strong negative themes
regular_positive_data = data2[data2['sentiment_score'] > 0].count()
regular_negative_data = data2[data2['sentiment_score'] < -0 ].count()
regular_neutral_data = data2[data2['sentiment_score'] == 0 ].count()
```

```python
import matplotlib.pyplot as plt

# Assuming you have already imported your data as 'data2'

# Calculate the counts of each sentiment category
positive_count = data2[data2['sentiment_score'] > 0].shape[0]
negative_count = data2[data2['sentiment_score'] < 0].shape[0]
neutral_count = data2[data2['sentiment_score'] == 0].shape[0]

# Calculate the percentages
total_count = len(data2)
positive_percentage = (positive_count / total_count) * 100
negative_percentage = (negative_count / total_count) * 100
neutral_percentage = (neutral_count / total_count) * 100

# Create a pie chart
labels = 'Positive', 'Negative', 'Neutral'
sizes = [positive_percentage, negative_percentage, neutral_percentage]
colors = ['lightgreen', 'lightcoral', 'grey']
explode = (0.1, 0.1, 0.1)  # To emphasize a particular slice, you can use 'explode'

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Sentiment Distribution')
plt.show()

```

```python
# targeting comments with strong positive themes
positive_data = data2[data2['sentiment_score'] > 0.3]
negative_data = data2[data2['sentiment_score'] < -0.3 ]
```

```python
text = " ".join(positive_data['content'])

from wordcloud import STOPWORDS

# Define the words you want to exclude
custom_stopwords = set(['flight', 'BA', 'Trip', 'Verified'])

# Add the custom stopwords to the default set of stopwords
stopwords = set(STOPWORDS).union(custom_stopwords)

# Create and generate the WordCloud with custom stopwords
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

# Display the WordCloud
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
import gensim
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis
import matplotlib.pyplot as plt

# Tokenization, stopwords removal, and lemmatization
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    tokens = word_tokenize(text)
    tokens = [word.lower() for word in tokens if word.lower() not in stop_words and word not in string.punctuation]
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return tokens

data2['processed_content'] = data2['content'].apply(preprocess_text)
```
